```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = df6.copy()
df_scaled = scaler.fit_transform(df6[df6._get_numeric_data().columns])
df_scaled = pd.DataFrame(df_scaled, columns=df6._get_numeric_data().columns)
df_scaled.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV
models_params = {
    'RandomForestClassifier': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100],  # Number of trees
            'max_depth': [None, 10, 20],     # Maximum depth of trees
            'min_samples_split': [2, 5, 10]  # Minimum samples to split a node
        }
    },
    'SVC': {
        'model': SVC(),
        'params': {
            'C': [1, 10],              # Regularization parameter
            'kernel': ['linear', 'rbf'],    # Kernel type
            'gamma': ['scale', 'auto']      # Kernel coefficient
        }
    },
    'KNeighborsClassifier': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [3, 5],       # Number of neighbors
            'weights': ['uniform', 'distance'],  # Weight function
            'metric': ['euclidean', 'manhattan'] # Distance metric
        }
    },
    'LogisticRegression': {
        'model': LogisticRegression(random_state=42),
        'params': {
            'penalty' : ['l1','l2'],
            'C' : [0.1, 0.5, 1.0 , 2.0],
            'solver' : ['liblinear', 'saga']
        }
        }
    }
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df_scaled.drop('Depression',axis=1),df_scaled['Depression'],test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def find_best_model(x,y,models_params):
  best_models = {}
  for model_name, mp in models_params.items():
    grid_search = GridSearchCV(mp['model'],mp['params'],cv=3,scoring='accuracy')
    grid_search.fit(x_train,y_train)
    best_models[model_name] = {
          'best_estimator': grid_search.best_estimator_,
          'best_params': grid_search.best_params_,
          'best_score': grid_search.best_score_
      }
  for name, model_info in best_models.items():
    best_model = model_info['best_estimator']
    y_pred = best_model.predict(x_test)
    test_accuracy = accuracy_score(y_test, y_pred)
    print(f"Model: {name}")
    print(f"Best Parameters: {model_info['best_params']}")
    print(f"Validation Accuracy (GridSearchCV): {model_info['best_score']:.4f}")
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print("-" * 50)
```

```python
find_best_model(x_train,y_train,models_params)
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='accuracy', cv=5, n_jobs=-1, verbose=2)

grid_search.fit(x_train, y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best accuracy score: ", grid_search.best_score_)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
test_accuracy = accuracy_score(y_test, y_pred)
print("Test accuracy: ", test_accuracy)
```
